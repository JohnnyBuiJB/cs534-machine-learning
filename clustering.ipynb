# Hierarchical Clustering

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
import numpy as np

def plot_HAC(X, y):
    fig, axs = plt.subplots(3, 1, figsize=(8,10), facecolor='w', edgecolor='k')
    axs = axs.ravel()

    clustering = AgglomerativeClustering(linkage='single').fit(X)
    axs[0].scatter(X[:, 0], X[:, 1], c=clustering.labels_, marker=".")
    axs[0].set_title("Single link")

    clustering = AgglomerativeClustering(linkage='complete').fit(X)
    axs[1].scatter(X[:, 0], X[:, 1], c=clustering.labels_, marker=".")
    axs[1].set_title("Complete link")

    clustering = AgglomerativeClustering(linkage='average').fit(X)
    axs[2].scatter(X[:, 0], X[:, 1], c=clustering.labels_, marker=".")
    axs[2].set_title("Average link")
    
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.3]])
blob_std = np.array([0.4, 0.3, 0.1])

X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)
plot_HAC(X, y)

X1, y1 = make_moons(n_samples=400, noise=0.05, random_state=28)
plot_HAC(X1, y1)
```

# K-means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2).fit(X1)
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans.labels_)
```

## K means through iterations

```python
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.3]])
blob_std = np.array([0.4, 0.3, 0.2])

X, y = make_blobs(n_samples=200, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)

fig, axs = plt.subplots(10, 1, figsize=(6,20), facecolor='w', edgecolor='k')

for i in range(0,10):
    kmeans = KMeans(n_clusters=2, max_iter=(i+1)).fit(X)
    axs[i].scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
```
